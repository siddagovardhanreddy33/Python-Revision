### 1. check if a number is even or odd

```python
n = int(input('enter a number : '))
print(n)

if n%2==0:
    print("even number")
else:
    print('obb number')
```

```python
# USING BITWISE OPERATOR

n = int(input("enter a number : "))
print(n)

if n & 1 ==0:
    print("Even number")
else:
    print("odd number")
```

```python
# USING TERNARY OPERATOR

n = int(input('enter a num : '))
print (n)

print('even' if n%2==0 else'odd')
```

```python
# USING FUNCTION

def even_odd_check(n):
    return 'even' if n%2==0 else 'odd'

even_odd_check(10)
```

```python
# USING LIST INDEXING

n = int(input('num : '))

result = ['even', 'odd']
print(result[n%2])
```

```python
# WITHOUT USING %, /, &, *

n = int(input('number :  '))
print(n)

while n>1:
    n-=2

if n==0:
    print('even')
else:
    print('odd')
```
